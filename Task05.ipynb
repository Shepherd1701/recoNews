```python
## 由于task04没有跑完，没有排序数据。自己另找一个MovieLens Latest Datasets Small数据集的emo跑一下
import pandas as pd
import numpy as np
```

```python
dtype = {"userId": np.int32, "movieId": np.int32, "rating": np.float32}
ratings = pd.read_csv('ml-latest-small/ratings.csv', dtype = dtype, usecols = range(3))
```

```python
# 构建透视表，找到用户和评分间的关系
ratings_matrix = ratings.pivot_table(index= ['userId'], columns=['movieId'], values = 'rating')
```

```python
similarity = ratings_matrix.T.corr()
# 工业上数据量太大，是不能这么算的
```

```python
## 预测用户1 和 物品1 的评分
similar_users = similarity[1].drop([1]).dropna()

# 找到正相关的用户
similar_users = similar_users.where(similar_users > 0).dropna()

# 在所有正相关的用户中找到所有消费过物品1的用户
ids = ratings_matrix[1].dropna().index & similar_users.index

similar_users.shape, idx.shape
```

```python
final_similar_users = similar_users.loc[list(ids)]
```

```python
## 计算用户1 对物品1的评分预测
sum_up = 0 # 保存评分预测的分母
sum_down = 0 # 保存分母
for sim_uid, similarity in final_similar_users.iteritems():
    sim_user_rated_movies = ratings_matrix.loc[sim_uid].dropna()
    sim_user_rating_for_item = sim_user_rated_movies[1]
    sum_up += similarity * sim_user_rating_for_item
    sum_down += similarity
predict_rating = sum_up/sum_down
```
